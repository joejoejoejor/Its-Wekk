```python
import pandas as pd
import numpy as np
```

```python
Parkhaus_total = pd.read_csv("./Parkh√§user_Traffic_Basel.csv", sep=";")
```

```python
Parkhaus_total.head()
```

```python
# Create a new dataframe containing only data for "Parking Elisabethen Basel"
elisabethen_df = Parkhaus_total[Parkhaus_total['Parkhaus'] == 'Parking Elisabethen Basel']

# Plot the data
elisabethen_df.plot(x='timestamp_text', y=['Einfahrten', 'Ausfahrten'], kind='line', title='Parking Elisabethen Basel Traffic')
# Convert timestamp_text to datetime
elisabethen_df['timestamp_text'] = pd.to_datetime(elisabethen_df['timestamp_text'])

# Extract year from timestamp_text
elisabethen_df['year'] = elisabethen_df['timestamp_text'].dt.year

# Plot data for each year
for year in elisabethen_df['year'].unique():
    yearly_data = elisabethen_df[elisabethen_df['year'] == year]
    yearly_data.plot(x='timestamp_text', y=['Einfahrten', 'Ausfahrten'], kind='line', title=f'Parking Elisabethen Basel Traffic {year}')
```

```python
# Filter data for the first week of January 2021
start_date = '2021-01-01'
end_date = '2021-01-07'
one_week_data = elisabethen_df[(elisabethen_df['timestamp_text'] >= start_date) & (elisabethen_df['timestamp_text'] <= end_date)]

# Display the filtered data
print(one_week_data)
one_week_data.plot(x='timestamp_text', y=['Einfahrten', 'Ausfahrten'], kind='line', title='Parking Elisabethen Basel Traffic (First Week of January 2021)')
```
