```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/Users/johannessolibieda/Documents/GitHub/Its-Wekk/4 - Data/4 - Johannes/100014.csv', delimiter=';')

# Display the first few rows of the DataFrame
df.head()
```

```python
# Get unique values in the 'Name' column
unique_names = df['Name'].unique()

# Create a dictionary to store DataFrames for each unique name
dfs = {name: df[df['Name'] == name] for name in unique_names}

# Example: Accessing the DataFrame for a specific name
# df_rebgasse = dfs['Rebgasse']
```

```python
for name in dfs:
    columns_to_select = ['Publikationszeit', 'Anzahl frei']
    dfs[name] = dfs[name][columns_to_select]
    print(f"DataFrame for {name}:")
    print(dfs[name].head())
```

```python
# Convert 'Publikationszeit' to datetime with utc=True
for name in dfs:
	dfs[name]['Publikationszeit'] = pd.to_datetime(dfs[name]['Publikationszeit'], utc=True)

# Round the 'Publikationszeit' to the nearest hour
for name in dfs:
	dfs[name]['RoundedTime'] = dfs[name]['Publikationszeit'].dt.round('H')

# Group by the rounded time and take the mean of 'Anzahl frei'
df_hourly = {name: dfs[name].groupby('RoundedTime')['Anzahl frei'].mean().reset_index() for name in dfs}

# Display the first few rows of the resulting DataFrame
for name in df_hourly:
	print(f"Hourly DataFrame for {name}:")
	print(df_hourly[name].head())
```

```python
from pandas.tseries.frequencies import to_offset

# Define the expected frequency
expected_freq = to_offset('H')

# Function to calculate gaps and display start and end of time series
def calculate_gaps_and_display_times(df):
    df = df.sort_values('RoundedTime')
    time_diff = df['RoundedTime'].diff()
    gaps = (time_diff > expected_freq).sum()
    start_time = df['RoundedTime'].min()
    end_time = df['RoundedTime'].max()
    return gaps, start_time, end_time

# Calculate gaps and display times for each DataFrame in dfs
gaps_and_times = {name: calculate_gaps_and_display_times(dfs[name]) for name in dfs}

# Print the number of gaps and start/end times for each DataFrame
for name in gaps_and_times:
    gaps, start_time, end_time = gaps_and_times[name]
    print(f"Number of gaps in the time series for {name}: {gaps}")
    print(f"Start of time series for {name}: {start_time}")
    print(f"End of time series for {name}: {end_time}")
```

```python
# Fill in time series gaps and interpolate 'Anzahl frei'
for name in df_hourly:
    df_hourly[name] = df_hourly[name].set_index('RoundedTime').resample('H').mean().interpolate().reset_index()

# Display the first few rows of the resulting DataFrame
for name in df_hourly:
    print(f"Interpolated Hourly DataFrame for {name}:")
    print(df_hourly[name].head())
```

```python
# Function to check for missing values in the time series
def check_missing_values(df):
    df = df.set_index('RoundedTime').resample('H').mean()
    missing_values = df['Anzahl frei'].isnull().sum()
    return missing_values

# Check for missing values in each DataFrame in df_hourly
missing_values = {name: check_missing_values(df_hourly[name]) for name in df_hourly}

# Print the number of missing values for each DataFrame
for name in missing_values:
    print(f"Number of missing values in the time series for {name}: {missing_values[name]}")
```

```python
# Filter each DataFrame in dfs for the entire day of January 1, 2022
for name in dfs:
    jan_first_data = dfs[name][(dfs[name]['RoundedTime'] >= '2022-01-01 00:00:00+00:00') & (dfs[name]['RoundedTime'] < '2022-01-02 00:00:00+00:00')]
    print(f"DataFrame for {name} on 2022-01-01:")
    print(jan_first_data)
```

```python
cutoff_time = pd.Timestamp('2022-01-01 00:00:00+0000', tz='UTC')

# Drop rows up to the cutoff time
for name in df_hourly:
    df_hourly[name] = df_hourly[name][df_hourly[name]['RoundedTime'] >= cutoff_time]

# Calculate the change to the next hour for each hour
for name in df_hourly:
    df_hourly[name]['Change'] = df_hourly[name]['Anzahl frei'].diff(-1).fillna(0)

# Drop the 'Anzahl frei' column
for name in df_hourly:
    df_hourly[name].drop(columns=['Anzahl frei'], inplace=True)

# Rename 'Change' column to include the name of the DataFrame
for name in df_hourly:
    df_hourly[name].rename(columns={'Change': f'{name}change'}, inplace=True)

# Display the first few rows of the resulting DataFrame
for name in df_hourly:
    print(f"Hourly DataFrame with Renamed Change for {name}:")
    print(df_hourly[name].head())
```

```python
from functools import reduce

# List of dataframes to merge
dfs_to_merge = [df_hourly[name] for name in df_hourly]

# Merge all dataframes on 'RoundedTime'
merged_df = reduce(lambda left, right: pd.merge(left, right, on='RoundedTime', how='outer'), dfs_to_merge)

# Display the first few rows of the merged dataframe
print(merged_df.head())
```
