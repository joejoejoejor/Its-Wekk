```python
import pandas as pd

df = pd.read_csv('/Users/johannessolibieda/Documents/GitHub/Its-Wekk/4 - Data/4 - Johannes/100051.csv', delimiter=';')  
df.head()
```

```python
df['Datum/Zeit'] = pd.to_datetime(df['timestamp_text'], format='%d.%m.%Y %H:%M')
df_sorted = df.sort_values(by='Datum/Zeit').reset_index(drop=True)
df_sorted.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Calculate the difference between consecutive timestamps
time_diff = df_sorted['Datum/Zeit'].diff()

# Check for gaps greater than 1 hour
gaps = time_diff[time_diff > pd.Timedelta(hours=1)]

print(gaps)
len(df_sorted)
```

```python
df_sorted = df_sorted.drop(columns=['EC [ug/m3]'])
df_sorted.head()
```

```python
df_sorted['TEMP [C]'] = df_sorted['TEMP [C]'].interpolate(method='linear')
df_sorted['PREC [mm]'] = df_sorted['PREC [mm]'].interpolate(method='linear')
df_sorted['RAD [W/m2]'] = df_sorted['RAD [W/m2]'].interpolate(method='linear')
df_sorted['CPC [1/cm3]'] = df_sorted['CPC [1/cm3]'].interpolate(method='linear')

# Fill missing values with the mean of values of similar weekday and time
df_sorted['weekday'] = df_sorted['Datum/Zeit'].dt.weekday
df_sorted['hour'] = df_sorted['Datum/Zeit'].dt.hour

for column in df_sorted.columns:
    if df_sorted[column].isnull().sum() > 0 and column not in ['Datum/Zeit', 'timestamp_text', 'weekday', 'hour']:
        df_sorted[column] = df_sorted.groupby(['weekday', 'hour'])[column].transform(lambda x: x.fillna(x.mean()))

df_sorted = df_sorted.drop(columns=['weekday', 'hour'])
```

```python
missing_values_updated = df_sorted.isnull().sum()
print(missing_values_updated)
```

```python
df_sorted.to_csv('1000051cleaned.csv', index=False)
```
