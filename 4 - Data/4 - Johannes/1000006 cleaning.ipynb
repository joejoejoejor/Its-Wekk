```python
import pandas as pd
import numpy as np

# Load the 100006 data file
df_100006 = pd.read_csv('/Users/johannessolibieda/Desktop/100006.csv', delimiter=';')

# Display the columns of the DataFrame to verify the structure
print(df_100006.columns)
```

```python
# Group the DataFrame by 'SiteCode'
grouped = df_100006.groupby('SiteCode')

# Create a dictionary to store DataFrames for each SiteCode
dfs_by_sitecode = {site_code: group for site_code, group in grouped}

# Sum up all values from 'PW' to 'andere' column for the same time slot and the same direction if there are multiple lanes
for site_code, df in dfs_by_sitecode.items():
    df_grouped = df.groupby(['DateTimeFrom', 'DateTimeTo', 'DirectionName']).agg({
        'PW': 'sum',
        'PW+': 'sum',
        'Lief': 'sum',
        'Lief+': 'sum',
        'Lief+Aufl.': 'sum',
        'LW': 'sum',
        'LW+': 'sum',
        'Sattelzug': 'sum',
        'Bus': 'sum',
        'andere': 'sum'
    }).reset_index()
    dfs_by_sitecode[site_code] = df_grouped

# Display the head of each DataFrame in the dictionary
for site_code, df in dfs_by_sitecode.items():
        print(f"SiteCode: {site_code}")
        display(df.head())
```

```python
# Check for missing values in each DataFrame in the dictionary
for site_code, df in dfs_by_sitecode.items():
    missing_values = df.isnull().sum()
    print(f"SiteCode: {site_code}")
    print(missing_values)
    print("\n")
```

```python
# Display the column names of each DataFrame in the dfs_by_sitecode dictionary
for site_code, df in dfs_by_sitecode.items():
    print(f"SiteCode: {site_code}, Columns: {df.columns.tolist()}")
```

```python
# Check if all DataFrames have the whole timespan filled
for site_code, df in dfs_by_sitecode.items():
    df['DateTimeFrom'] = pd.to_datetime(df['DateTimeFrom'], utc=True)
    df['DateTimeTo'] = pd.to_datetime(df['DateTimeTo'], utc=True)
    full_timespan = pd.date_range(start=df['DateTimeFrom'].min(), end=df['DateTimeTo'].max(), freq='h', tz='UTC')
    missing_times = full_timespan.difference(df['DateTimeFrom'])
    print(f"SiteCode: {site_code}")
    if missing_times.empty:
        print("All timeslots are filled.")
    else:
        print(f"Missing timeslots: {missing_times}")
    print(f"Number of missing timeslots: {len(missing_times)}")
    print("\n")
```

```python
from pandas.tseries.frequencies import to_offset

# Function to fill missing time slots with mean values
def fill_missing_timeslots(df, full_timespan):
    # Remove duplicates based on DateTimeFrom
    df = df.drop_duplicates(subset='DateTimeFrom')
    
    # Set DateTimeFrom as the index
    df = df.set_index('DateTimeFrom')
    
    # Reindex the DataFrame to include the full timespan
    df = df.reindex(full_timespan)
    
    # Reset the index
    df = df.reset_index().rename(columns={'index': 'DateTimeFrom'})
    
    # Fill missing values with the mean of the same time on the same weekday
    df['Weekday'] = df['DateTimeFrom'].dt.weekday
    df['Hour'] = df['DateTimeFrom'].dt.hour
    for column in ['PW', 'PW+', 'Lief', 'Lief+', 'Lief+Aufl.', 'LW', 'LW+', 'Sattelzug', 'Bus', 'andere']:
        df[column] = df.groupby(['Weekday', 'Hour'])[column].transform(lambda x: x.fillna(x.mean()))
    
    # Drop the Weekday and Hour columns
    df = df.drop(columns=['Weekday', 'Hour'])
    
    return df

# Apply the function to each DataFrame in the updated_dfs_by_direction dictionary
for key, df in updated_dfs_by_direction.items():
    full_timespan = pd.date_range(start=df['DateTimeFrom'].min(), end=df['DateTimeFrom'].max(), freq='h', tz='UTC')
    updated_dfs_by_direction[key] = fill_missing_timeslots(df, full_timespan)

# Display the keys of the updated dictionary to verify
print(updated_dfs_by_direction.keys())
```

```python
# Drop 'DateTimeTo' and 'DirectionName' columns from each DataFrame in the updated_dfs_by_direction dictionary
for key, df in updated_dfs_by_direction.items():
    columns_to_drop = [col for col in ['DateTimeTo', 'DirectionName'] if col in df.columns]
    updated_dfs_by_direction[key] = df.drop(columns=columns_to_drop)

# Rename all columns with the name of df + column name except the first column
for key, df in updated_dfs_by_direction.items():
    df.columns = [df.columns[0]] + [f"{key}{col}" for col in df.columns[1:]]

# Display the column names of each DataFrame to verify
for key, df in updated_dfs_by_direction.items():
    print(f"Key: {key}, Columns: {df.columns.tolist()}")
```

```python
# Function to reduce the number of columns and keep the site code in front of the new column names
def reduce_columns(df, key):
    df[f'{key}_sumPW'] = df[f'{key}PW'] + df[f'{key}PW+']
    df[f'{key}_sumLief'] = df[f'{key}Lief'] + df[f'{key}Lief+'] + df[f'{key}Lief+Aufl.']
    df[f'{key}_sumLW'] = df[f'{key}LW'] + df[f'{key}LW+'] + df[f'{key}Sattelzug'] + df[f'{key}Bus'] + df[f'{key}andere']
    return df[['DateTimeFrom', f'{key}_sumPW', f'{key}_sumLief', f'{key}_sumLW']]

# Apply the function to each DataFrame in the updated_dfs_by_direction dictionary
for key, df in updated_dfs_by_direction.items():
    updated_dfs_by_direction[key] = reduce_columns(df, key)

# Display the column names of each DataFrame to verify
for key, df in updated_dfs_by_direction.items():
    print(f"Key: {key}, Columns: {df.columns.tolist()}")
```

```python
# Find the latest common end time for all dataframes
latest_common_end_time = min(df['DateTimeFrom'].max() for df in updated_dfs_by_direction.values())

# Filter each dataframe to include only rows up to the latest common end time
filtered_dfs = {key: df[df['DateTimeFrom'] <= latest_common_end_time] for key, df in updated_dfs_by_direction.items()}

# Merge all datasets on 'DateTimeFrom'
merged_df = pd.concat([df.set_index('DateTimeFrom') for df in filtered_dfs.values()], axis=1).reset_index()

# Display the merged DataFrame
print(merged_df.head())

# Display the size of the merged DataFrame
print(f"Size of the merged DataFrame: {merged_df.size}")

# Display the number of columns and rows of the merged DataFrame
print(f"Number of columns: {merged_df.shape[1]}")
print(f"Number of rows: {merged_df.shape[0]}")
```
