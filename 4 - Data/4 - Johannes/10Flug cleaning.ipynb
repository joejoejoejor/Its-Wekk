```python
import pandas as pd

file_path = '/Users/johannessolibieda/Documents/GitHub/Its-Wekk/4 - Data/4 - Johannes/Flugbewegungen_cleaned_daily.csv'
df = pd.read_csv(file_path, delimiter=';')
df.head()
```

```python
import pandas as pd

# Create a new dataframe with a row for each hour of the day for each date
full_day_df = df.loc[df.index.repeat(24)].reset_index(drop=True)
full_day_df['Hour'] = full_day_df.groupby('Datum').cumcount()

# Set 'Flugbewegungen_per_hour' for each hour
full_day_df['Flugbewegungen_per_hour'] = full_day_df.apply(
    lambda row: row['Flugbewegungen_per_hour'] if 5 <= row['Hour'] < 24 else 0, axis=1
)

# Combine 'Datum' and 'Hour' into a single datetime column
full_day_df['Datum'] = full_day_df.apply(
    lambda row: row['Datum'] + pd.Timedelta(hours=row['Hour']), axis=1)

# Display the updated dataframe
full_day_df.head(48)
```

```python
hour_df = full_day_df

# Check if 'Hour' column exists before dropping it
if 'Hour' in hour_df.columns:
	hour_df.drop(['Hour'], axis=1, inplace=True)

# Check if 'Flugbewegungen' column exists before dropping it
if 'Flugbewegungen' in hour_df.columns:
	hour_df.drop(['Flugbewegungen'], axis=1, inplace=True)
```
