## Autocorrelation and Partial Autocorrelation Functions

We use the autocorrelation function (ACF) and partial autocorrelation function (PACF) to study the serial dependence properties of a time series and guide its modeling. Different time series models exhibit distinct patterns in the ACF and PACF, which can help us determine an appropriate model structure.

**Observation 1**
Both the ACF and PACF start with a lag of 0, where the partial autocorrelation and autocorrelation are 1. 

**Observation 2**
|                | ACF $\rho(k)$                                         | PACF $\phi_{kk}$                                      |
|----------------|-------------------------------------------------------|-------------------------------------------------------|
| **AR(p)**      | Damped exponential and/or sine functions              | $\phi_{kk} = 0$ for $k > p$                           |
| **MA(q)**      | $\rho(k) = 0$ for $k > q$                             | Dominated by damped exponential and/or sine functions |
| **ARMA(p, q)** | Damped exponential and/or sine functions after lag $\max(0, q - p)$ | Dominated by damped exponential and/or sine functions after lag $\max(0, p - q)$ |

```python
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

# Use a nicer style for plots
plt.style.use("seaborn-v0_8-muted")
```

```python
np.random.seed(0)  # Set a random seed for reproducibility
# TODO Change to non-stationary time-series
epsilon_t = np.random.normal(0, 1, 1000) + np.arange(0, 1000)

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
sm.graphics.tsa.plot_acf(epsilon_t.squeeze(), lags=40, ax=ax[0], auto_ylims=True)
sm.graphics.tsa.plot_pacf(epsilon_t.squeeze(), lags=40, ax=ax[1], auto_ylims=True)
plt.show()
```
