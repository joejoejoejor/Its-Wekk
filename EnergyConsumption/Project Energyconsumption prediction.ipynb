```python
import numpy as np
import pandas as pd
```

```python
years13till15 = pd.read_csv("./Energieverbrauch_Winterthur_2013-15.csv", sep=";", decimal=",")
years16till18 = pd.read_csv("./Energieverbrauch_Winterthur_2016-18.csv", sep=";", decimal=",")
years19till21 = pd.read_csv("./Energieverbrauch_Winterthur_2019-21.csv", sep=";", decimal=",")
yearsfrom2022 = pd.read_csv("./Energieverbrauch_Winterthur_ab_2022.csv", sep=";", decimal=",")
```

```python
years13till15.head(20)
```

```python
import matplotlib.pyplot as plt

# Split the column into two separate columns
years13till15[['zeitpunkt', 'bruttolastgang_kwh']] = years13till15['zeitpunkt,"bruttolastgang_kwh"'].str.split(',', expand=True)

# Convert the columns to appropriate data types
years13till15['zeitpunkt'] = pd.to_datetime(years13till15['zeitpunkt'])
years13till15['bruttolastgang_kwh'] = pd.to_numeric(years13till15['bruttolastgang_kwh'])

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(years13till15['zeitpunkt'], years13till15['bruttolastgang_kwh'])
plt.xlabel('Zeitpunkt')
plt.ylabel('Bruttolastgang (kWh)')
plt.title('Energieverbrauch Winterthur 2013-2015')
plt.show()

# Split the column into two separate columns for years16till18
years16till18[['zeitpunkt', 'bruttolastgang_kwh']] = years16till18['zeitpunkt,"bruttolastgang_kwh"'].str.split(',', expand=True)

# Convert the columns to appropriate data types for years16till18
years16till18['zeitpunkt'] = pd.to_datetime(years16till18['zeitpunkt'])
years16till18['bruttolastgang_kwh'] = pd.to_numeric(years16till18['bruttolastgang_kwh'])

# Plot the data for years16till18
plt.figure(figsize=(10, 6))
plt.plot(years16till18['zeitpunkt'], years16till18['bruttolastgang_kwh'])
plt.xlabel('Zeitpunkt')
plt.ylabel('Bruttolastgang (kWh)')
plt.title('Energieverbrauch Winterthur 2016-2018')
plt.show()

# Split the column into two separate columns for years19till21
years19till21[['zeitpunkt', 'bruttolastgang_kwh']] = years19till21['zeitpunkt,"bruttolastgang_kwh"'].str.split(',', expand=True)

# Convert the columns to appropriate data types for years19till21
years19till21['zeitpunkt'] = pd.to_datetime(years19till21['zeitpunkt'])
years19till21['bruttolastgang_kwh'] = pd.to_numeric(years19till21['bruttolastgang_kwh'])

# Plot the data for years19till21
plt.figure(figsize=(10, 6))
plt.plot(years19till21['zeitpunkt'], years19till21['bruttolastgang_kwh'])
plt.xlabel('Zeitpunkt')
plt.ylabel('Bruttolastgang (kWh)')
plt.title('Energieverbrauch Winterthur 2019-2021')
plt.show()

# Split the column into two separate columns for yearsfrom2022
yearsfrom2022[['zeitpunkt', 'bruttolastgang_kwh']] = yearsfrom2022['zeitpunkt,"bruttolastgang_kwh"'].str.split(',', expand=True)

# Convert the columns to appropriate data types for yearsfrom2022
yearsfrom2022['zeitpunkt'] = pd.to_datetime(yearsfrom2022['zeitpunkt'])
yearsfrom2022['bruttolastgang_kwh'] = pd.to_numeric(yearsfrom2022['bruttolastgang_kwh'])

# Plot the data for yearsfrom2022
plt.figure(figsize=(10, 6))
plt.plot(yearsfrom2022['zeitpunkt'], yearsfrom2022['bruttolastgang_kwh'])
plt.xlabel('Zeitpunkt')
plt.ylabel('Bruttolastgang (kWh)')
plt.title('Energieverbrauch Winterthur ab 2022')
plt.show()


# Compute the total energy consumption per year for each dataset
total_energy_13_15 = years13till15.groupby(years13till15['zeitpunkt'].dt.year)['bruttolastgang_kwh'].sum()
total_energy_16_18 = years16till18.groupby(years16till18['zeitpunkt'].dt.year)['bruttolastgang_kwh'].sum()
total_energy_19_21 = years19till21.groupby(years19till21['zeitpunkt'].dt.year)['bruttolastgang_kwh'].sum()
total_energy_from_2022 = yearsfrom2022.groupby(yearsfrom2022['zeitpunkt'].dt.year)['bruttolastgang_kwh'].sum()

# Print the total energy consumption per year
print("Total energy consumption per year (2013-2015):")
print(total_energy_13_15)
print("\nTotal energy consumption per year (2016-2018):")
print(total_energy_16_18)
print("\nTotal energy consumption per year (2019-2021):")
print(total_energy_19_21)
print("\nTotal energy consumption per year (from 2022):")
print(total_energy_from_2022)
```

```python

# Combine the data into a single DataFrame
data = pd.concat([years13till15, years16till18, years19till21, yearsfrom2022])
```

```python
years13till15.head(20)
```

```python
# Convert the 'zeitpunkt' column to datetime format
data['zeitpunkt'] = pd.to_datetime(data['zeitpunkt'])

# Sort the data by date
data = data.sort_values(by='zeitpunkt')
```

```python


import matplotlib.pyplot as plt

plt.plot(data['zeitpunkt'], data['bruttolastgang_kwh'])
plt.title('Energy Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Consumption (kWh)')
plt.show()
```

```python


from statsmodels.tsa.stattools import adfuller

result = adfuller(data['bruttolastgang_kwh'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plot_acf(data['bruttolastgang_kwh'])
plot_pacf(data['bruttolastgang_kwh'])
plt.show()
```
